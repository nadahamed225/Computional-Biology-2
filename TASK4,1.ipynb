```python
import networkx as nx
import pandas as pd
```

```python
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4])
G.add_edges_from([(1,2),(1,3),(2,3),(3,4)])
```

```python
nx.draw(G,with_labels=True,node_size=2000)
```

```python
list(G.neighbors(1))
```

```python
for neighbor in G.neighbors(1):
    print(neighbor)
```

```python
G.has_node(2)
```

```python
nx.is_tree(G)
```

```python
nx.is_connected(G)
```

```python
G.degree(1)
```

```python
D = nx.DiGraph()
D.add_edges_from([(1,2), (1,3), (2,4), (3,4)])
nx.draw(D, with_labels = True)
```

```python
X=nx.Graph()
X.add_edges_from([(1,2),(2,3),(1,3),(3,4)])
nx.draw(X,with_labels=True,node_size=500)
```
